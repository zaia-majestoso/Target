```python
from random import randint
import time
import math
import numpy as np
import psutil as ps
import tracemalloc



#def begin():
  #tracemalloc.start()
  #start = time.time_ns()
  #return start

#def stop():
  #current, peak = tracemalloc.get_traced_memory()
  #end = time.time_ns()
  #tracemalloc.stop()
  #return end, current


def Criavet():
  start = time.time_ns()
  tracemalloc.start()
  vet = []
  for num in range(1, 1000):
    vet.append(randint(0, 10000))
  end = time.time_ns()
  current, peak = tracemalloc.get_traced_memory()
  tracemalloc.stop()
  print(f"Criavet() memory usage: {current / 1024}KB")
  print(f"tempo para criar o vetor: {end - start} nanosegundos")
  
  

  ord(vet)
  imprime(vet)


def ord(vet):
  tracemalloc.start()
  start = time.time_ns()
  vet.sort()
  current, peak = tracemalloc.get_traced_memory()
  tracemalloc.stop()
  end = time.time_ns()
  print(f"tempo para ordenar o vetor: {end - start} nanosegundos")
  print(f"ord(vet) memory usage: {current / 1024}KB")



def imprime(vet):
  tracemalloc.start()
  start = time.time_ns()
  print(vet)
  current, peak = tracemalloc.get_traced_memory()
  tracemalloc.stop()
  end = time.time_ns()
  print(f"tempo para imprimir o vetor: {end - start} nanosegundos")
  print(f"imprime(vet) memory usage: {current / 1024}KB")

Criavet()


def pi100(num):
  pi = 0.0
  sinal = 1.0
  for i in range(num):
    pi += sinal/(2.0 * i + 1.0)
    sinal = -sinal
  pi *= 4.0
  return pi

pi = pi100(10000)
print("{:.100f}".format(pi))
```
