# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Definindo as funções
def f(x):
    return x * np.exp(-x**2)

def g(x):
    return x + 1

# Função para a qual queremos encontrar as raízes
def h(x):
    return f(x) - g(x)

# Encontrando o ponto de interseção das duas funções usando o método de Newton
x_intersection = newton(h, 0)

# Definindo o intervalo de x
x = np.linspace(-1, 2, 1000)

# Calculando os valores de y
y1 = f(x)
y2 = g(x)

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Plotando as funções
ax.plot(x, y1, 'r', label='y = xe^-x^2')
ax.plot(x, y2, 'b', label='y = x + 1')

# Preenchendo a área entre as curvas
ax.fill_between(x, y1, y2, where=(x >= x_intersection) & (x <= 2), color='gray', alpha=0.5)

# Adicionando uma linha vertical em x = 2
ax.axvline(x=2, color='k', linestyle='--')

# Adicionando legendas
ax.legend()

# Mostrando o gráfico
plt.show()

```

Encontre o valor da constante C para a antiderivada $F(x)$ da função $f(x) = 3x^3 - 2x + 1$ de tal forma que $F(1)=0$. Apresente seus cálculos em latex e explique as decisões que tomou.

```python
from sympy import symbols, integrate, solve
from IPython.display import display, Math

# Definindo a variável simbólica
x, C = symbols('x C')

# Definindo a função
f = 3*x**2 - 2*x + 1

# Calculando a integral indefinida de f
F = integrate(f, x) + C
display(Math('F(x) = C + x^3 - x^2 + x'))

# Encontrando a constante de integração C tal que F(1) = 0
# C sera -1 para satisfazer a condição apresentada
C_value = solve(F.subs(x, 1), C)[0]
display(Math('C = ' + str(C_value)))

# A função F com a constante de integração encontrada
F = F.subs(C, C_value)
display(Math('F(x) = x^3 - x^2 + x - 1'))
```
